```python
# Bronze Layer - Read from FileStore and save as Delta tables
from pyspark.sql.functions import current_timestamp

# ── FMCG Sales ──────────────────────────────────────────
fmcg_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Volumes/retail_shelf_intelligence/raw_data/retail_shelf_raw/Indian FMCG Retail Sales  Customer  Inventory (2024).csv")

fmcg_df = fmcg_df.withColumn("ingested_at", current_timestamp())

fmcg_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("retail_shelf_intelligence.raw_data.bronze_fmcg_sales")

print(f"Bronze FMCG Sales saved: {fmcg_df.count()} rows")

# ── Weather Data ─────────────────────────────────────────
weather_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Volumes/retail_shelf_intelligence/raw_data/retail_shelf_raw/weather_2024.csv")

weather_df = weather_df.withColumn("ingested_at", current_timestamp())

weather_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("retail_shelf_intelligence.raw_data.bronze_weather")

print(f"Bronze Weather saved: {weather_df.count()} rows")

# ── Holidays Data ─────────────────────────────────────────
holidays_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Volumes/retail_shelf_intelligence/raw_data/retail_shelf_raw/holidays_2024.csv")

holidays_df = holidays_df.withColumn("ingested_at", current_timestamp())

holidays_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("retail_shelf_intelligence.raw_data.bronze_holidays")

print(f"Bronze Holidays saved: {holidays_df.count()} rows")

print("\n✅ Bronze layer complete!")
```
